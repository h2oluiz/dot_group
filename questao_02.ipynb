```python
from langchain_openai import ChatOpenAI
from langchain.prompts import ChatPromptTemplate
from langchain.memory import ChatMessageHistory
from langchain_core.runnables import RunnableWithMessageHistory
```

```python
# prompt
template = """
Voc√™ √© assistente especializado exclusivamente \
em responder perguntas sobre a linguagem Python.
Se sua pergunta n√£o for relacionada a Python, \
informarei que s√≥ posso ajudar com t√≥picos dessa \
linguagem e pedirei que voc√™ a reformule.
Usu√°rio: {pergunta}
Programador:"""

# Prompt Template
prompt = ChatPromptTemplate.from_template(template)

# llm gpt-5
llm = ChatOpenAI(model="gpt-5-nano", temperature=0.7)

# Chain b√°sica
chain = prompt | llm

# Store de mem√≥rias
store = {}

def get_history(session_id: str):
    """Retorna ou cria o hist√≥rico da sess√£o."""
    if session_id not in store:
        store[session_id] = ChatMessageHistory()
    return store[session_id]

# Cadeia com mem√≥ria ativada
chatbot = RunnableWithMessageHistory(
    chain,
    get_history,
    input_messages_key="pergunta",
)
```

```python
print("üêç Assistente especialista em Python! \
Digite sua pergunta ou 'sair' para encerrar. üêç\n")

session_id = "luiz_santos-01"

while True:
    user_input = input("Voc√™: ")

    if user_input.lower() == "sair":
        print("Fui!!!!. At√© mais!")
        break

    response = chatbot.invoke(
        {"pergunta": user_input},
        config={"session_id": session_id}
    )
    print("\nProgramador üêç :", response.content)
    print("=" * 80)
```
